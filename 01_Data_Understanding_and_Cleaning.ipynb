Phase 3 – Feature Engineering
Objective

Transform raw transactional data into business-ready data for analytics, SQL, and Power BI.


```python
# Year
df["Year"] = df["Order Date"].dt.year

# Quarter
df["Quarter"] = df["Order Date"].dt.quarter

# Month Number
df["Month"] = df["Order Date"].dt.month

# Month Name
df["Month Name"] = df["Order Date"].dt.strftime("%B")

# Day of Week
df["Day"] = df["Order Date"].dt.day_name()
```

```python
df["Shipping Days"] = (
    df["Ship Date"] - df["Order Date"]
).dt.days
```

```python
df["Profit Margin %"] = (
    df["Profit"] / df["Sales"]
) * 100
df["Profit Margin %"] = df["Profit Margin %"].round(2)
```

```python
df["Loss Order"] = np.where(
    df["Profit"] < 0,
    "Yes",
    "No"
)
```

```python
def discount_category(x):

    if x == 0:
        return "No Discount"

    elif x <= 0.20:
        return "Low"

    elif x <= 0.40:
        return "Medium"

    else:
        return "High"


df["Discount Category"] = df["Discount"].apply(discount_category)
```

```python
def sales_category(x):

    if x < 100:
        return "Low"

    elif x < 500:
        return "Medium"

    elif x < 1000:
        return "High"

    else:
        return "Very High"


df["Sales Category"] = df["Sales"].apply(sales_category)
```

```python
def order_size(x):

    if x == 1:
        return "Single Item"

    elif x <= 5:
        return "Small Order"

    elif x <= 10:
        return "Medium Order"

    else:
        return "Bulk Order"


df["Order Size"] = df["Quantity"].apply(order_size)
```
